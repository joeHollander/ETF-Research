```python
import databento as db
import os
import pandas as pd
import numpy as np
```

```python
client = db.Historical(key=os.getenv("DATABENTO_API_KEY"))
```

```python
params = {
    "dataset": "GLBX.MDP3",
    "symbols": "GC.c.0, GC.c.1, GC.c.2, GC.c.6, GC.c.12",
    "stype_in": "continuous",
    "schema": "ohlcv-1d",
    "start": "2010-06-06",
    "end": "2025-08-18"
}

cost = client.metadata.get_cost(**params)
print(f"Estimated cost for this download: ${cost:.2f}")
```

```python
df = pd.read_csv(r"Data\gold_continuous_futures.csv")
```

```python
mini_df = df[df['symbol'] == 'GC.c.0'].copy()
last_day = mini_df['instrument_id'] != mini_df['instrument_id'].shift(-1)
print(f"Average volume on contract rollover days: {mini_df.loc[last_day, 'volume'].mean():.2f}")
```

```python
roll_adj = mini_df.loc[last_day, 'open'].shift(-1) - mini_df.loc[last_day, 'close']
print(f"Average roll adjustment on contract rollover days: {roll_adj.mean():.2f}")
```

```python
gen_adj = mini_df['open'].shift(-1) - mini_df['close']
print(f"Average daily adjustment: {gen_adj.mean()}")
```
